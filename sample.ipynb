```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from utils import check, create
```

```python
generator = load_model("./models/generator64.h5")
```

```python
imgs = [create(generator, size=(128,128)) for i in tqdm(range(20))]
```

```python
row=4
col=5
plt.figure(figsize=(col*5, row*5))
for i in range(row * col):
    plt.subplot(row, col, i+1)
    plt.imshow(imgs[i])
    plt.axis("off")
```
